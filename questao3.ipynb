```python
from google.cloud import bigquery
from google.colab import auth
import pandas as pd
import numpy as np
```

```python
# Authenticate with Google Cloud
auth.authenticate_user()
```

```python
# Construct a BigQuery client object.
client = bigquery.Client(project='analise-trabalhadores') # Tem que procurar o seu ID de projeto pessoal no GCP
```

```python
query = """
SELECT *
FROM basedosdados.br_me_rais.microdados_estabelecimentos
WHERE ano BETWEEN 2019 AND 2023 AND indicador_atividade_ano = 1
"""
```

```python
result = client.query(query)
```
